```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
df = pd.read_csv('data/PS_part_aa.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
# check data types
df.dtypes
```

```python
# numeric features
numeric_features = ['step', 'amount',
                    'oldbalanceOrg', 'newbalanceOrig',
                    'oldbalanceDest', 'newbalanceDest']

# categorical features
categorical_features = ['type', 'nameOrig', 'nameDest']
```

```python
# Missing values
df.isna().sum()
```

```python
df.isna()
```

```python
df = df.dropna()
```

```python
# Imputation
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(fill_value=np.nan, strategy='mean')
imputed_data = imputer.fit_transform(df[numeric_features])
imputed_df = pd.DataFrame(imputed_data)
df[numeric_features] = imputed_df
```

```python
imputed_df
```

```python
df
```

```python
# Correlation between x and y
y_column = 'isFraud'
x_column = 'type'
df['isFraud'].mean() # -> 0.13%
```

```python
display(pd.crosstab(index=df[x_column], columns=df[y_column], normalize='columns'))
```

```python
# Histograms for each numeric features
display(df.describe())
hist = df.hist(bins=30, sharey=True, figsize=(10, 10))
```

```python
# Histogram
for column in df[numeric_features].columns:    
    print(column)
    hist = df[column].hist(by=df[y_column], bins=30)
    plt.show()
```

```python
# numeric feature: scaling
scaler = MinMaxScaler()
transformed_data = scaler.fit_transform(df[numeric_features])
```

```python
transformed_data
```

```python
transformed_df = pd.DataFrame(transformed_data)
transformed_df
```

```python
# categorical feature: one-hot encoding
df[['type']].head(10)
```

```python
one_hot = pd.get_dummies(df['type'])
one_hot.head(10)
```

```python
# combine original + numeric + categorical features
final_df = pd.concat([df, transformed_df, one_hot], axis = 1)
```
